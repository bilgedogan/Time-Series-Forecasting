```python
def plot_histogram(y,x,train,column_names,colors):
    plt.figure(figsize=(20,20))
    for i in range(train.shape[1]):
        plt.subplot(y,x,i+1)
        plt.hist(train[:,i],color=colors[i])
        plt.title(col_names[i])

    plt.show()
```

### Smoothing - Moving Average

```python
def moving_average(observations, window):
    cumulative_sum = np.cumsum(observations, dtype=float)
    cumulative_sum[window:] = cumulative_sum[window:] - cumulative_sum[:-window]
    return cumulative_sum[window - 1:] / window
```

```python
for i in range(train.shape[1]):
    window = 100
    moving_avg_train = moving_average(train[:,i], window=window)
    plt.plot(train[:1000,i],color="black")
    plt.plot(moving_avg_train[:1000],color="red")
    plt.title(col_names[i])
    plt.legend(["train","moving_avg"])
    plt.show()
    
    print("RMSE for {}:{}".format(col_names[i], mean_squared_error(train[window-1:,i],moving_avg_train)))
```

### Single Exponential Smoothing

```python
from statsmodels.tsa.api import SimpleExpSmoothing

for i in range(train.shape[1]):

    single = SimpleExpSmoothing(train[:,i]).fit(optimized=True)
    single_preds = single.forecast(len(train[:,i]))
    
    plt.plot(train[:,i],color="black")
    plt.plot(single_preds,color="red")
    plt.title(col_names[i])
    plt.legend(["train","single_preds"])
    plt.show()
    print("RMSE for {}:{}".format(col_names[i], mean_squared_error(train[:,i],single_preds)))
```

### Double Exponential Smoothing

```python
from statsmodels.tsa.api import Holt

for i in range(train.shape[1]):

    double = Holt(train[:,i]).fit(optimized=True)
    double_preds = single.forecast(len(train[:,i]))
    
    plt.plot(train[:,i],color="black")
    plt.plot(double_preds,color="red")
    plt.title(col_names[i])
    plt.legend(["train","double_preds"])
    plt.show()
    print("RMSE for {}:{}".format(col_names[i], mean_squared_error(train[:,i],double_preds)))
```

### Triple Exponential Smoothing

```python
from statsmodels.tsa.api import ExponentialSmoothing

for i in range(train.shape[1]):
    triple = ExponentialSmoothing(train[:,i],
                              trend="additive").fit(optimized=True)
    triple_preds = triple.forecast(len(train[:,i]))
    plt.plot(train[:,i],color="black")
    plt.plot(triple_preds,color="red")
    plt.title(col_names[i])
    plt.legend(["train","double_preds"])
    plt.show()
    print("RMSE for {}:{}".format(col_names[i], mean_squared_error(train[:,i],triple_preds)))
```

```python
for index in range(train.shape[1]):
    dt_feature = [item for item in train[:,index]]
    first_quartile,third_quartile = np.percentile(dt_feature , [25,75])
    iqr = third_quartile-first_quartile
    upper_limit = third_quartile+1.5*iqr
    lower_limit = first_quartile-1.5*iqr
    outlier_indices = [i for i in range(train.shape[0]) if (dt_feature[i]>upper_limit or dt_feature[i]<lower_limit) ]
    
    
    outlier_val = []
    for i in range(len(outlier_indices)):
        outlier_val.append(dt_feature[outlier_indices[i]])

    plt.plot(dt_feature,color='b')
    plt.title(col_names[index])
    plt.scatter(outlier_indices,outlier_val,color='r')
    plt.legend(['original','anomaly'], loc='upper right')
    plt.show()
    
    
```
